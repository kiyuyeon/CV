```python
src = cv2.imread('./Ch_2_fig/fig/airplane.bmp')
mask = cv2.imread('./Ch_2_fig/fig/mask_plane.bmp')
dst = cv2.imread('./Ch_2_fig/fig/field.bmp')


if src is None or mask is None or dst is None:
    print('image read failed')
    sys.exit()

cv2.copyTo(src=src,mask=mask,dst=dst)

cv2.imshow('src',src)
cv2.imshow('mask',mask)
cv2.imshow('dst',dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import os 
path_lists = []
paths = os.listdir('./Ch_1_fig/fig/images/')
# print(paths)
for i in paths:
    path_list = './Ch_1_fig/fig/images/'+ i
    path_lists.append(path_list)

path_lists

cv2.namedWindow('scene',cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene',cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN)

idx=0
while True:
    img = cv2.imread(path_lists[idx])

    if img is None:
        continue

    cv2.imshow('scene',img)

    if cv2.waitKey(3000)==27:
        break

    idx += 1

    if idx >=len(path_lists):
        idx=0

cv2.destroyAllWindows()
```

```python
img_BGR = cv2.imread('./Ch_1_fig/fig/puppy.bmp',cv2.IMREAD_COLOR)
img_RGB = cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)
img_GRAY = cv2.cvtColor(img_BGR,cv2.COLOR_BGR2GRAY)

fig ,ax =plt.subplots(1,3,figsize=(12,5))

ax[0].imshow(img_BGR),ax[0].axis('off'),ax[0].set_title('BGR')
ax[1].imshow(img_RGB),ax[1].axis('off'),ax[1].set_title('RGB')
ax[2].imshow(img_GRAY,cmap='gray'),ax[2].axis('off'),ax[2].set_title('GRAY')

plt.show()
```

```python
src = cv2.imread('./Ch_2_fig/fig/cow.png')
dst = cv2.imread('./Ch_2_fig/fig/green.png')

##MASK
src_gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
cv2.threshold(src,150,255,cv2.THRESH_BINARY)

_,maks=cv2.threshold(src_gray,240,255,cv2.THRESH_BINARY_INV)

h,w= src.shape[:2]
# dst = cv2.resize(dst,(w,h),interpolation=cv2.INTER_AREA)
dst_crop = dst[400:400*h,400:400+w]

cv2.copyTo(src,maks,dst_crop)

# cv2.imshow('src',src)
# cv2.imshow('src_gray',src_gray)
cv2.imshow('mask',maks)
cv2.imshow('dst_crop',dst_crop)


cv2.waitKey()
cv2.destroyAllWindows()
```
